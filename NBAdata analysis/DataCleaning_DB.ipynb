```python
import numpy as np
import pandas as pd
```

# Data Cleaning for game details csv

```python
games_details_data=pd.read_csv('./games_details.csv',)

games_details_data
```

```python
percent_missing = games_details_data.isnull().sum() * 100 / len(games_details_data)
percent_missing

games_details_data.drop(['NICKNAME','START_POSITION','COMMENT'],axis=1,inplace= True)
games_details_data.dropna(inplace=True)
games_details_data.drop_duplicates(subset=['GAME_ID','TEAM_ID','PLAYER_ID'],inplace=True)
games_details_data.to_csv('./game_details_modified.csv',index=False)
```

# Create players.csv using game details

```python
players = games_details_data[['PLAYER_ID','PLAYER_NAME']]
players.drop_duplicates(subset=['PLAYER_ID'],inplace=True)
players.to_csv('./players.csv',index=False)
```

#  Data Cleaning for game csv

```python
games_data=pd.read_csv('./games.csv')

percent_missing_games = games_data.isnull().sum() * 100 / len(games_data)


unique_games=games_data.drop_duplicates(subset='GAME_ID', keep="first")
print(games_data.nunique())
percent_missing_games
unique_games.to_csv('./games_modified.csv',index=False)
games_data.drop(['GAME_STATUS_TEXT','TEAM_ID_home','VISITOR_TEAM_ID','FG_PCT_home','FG_PCT_away'],axis=1).to_csv('./games_modified.csv',index=False)
```

#  Data Cleaning for ranking csv

```python
ranking_data=pd.read_csv('./ranking.csv')

percent_missing_ranking = ranking_data.isnull().sum() * 100 / len(ranking_data)
percent_missing_ranking
ranking_data.drop_duplicates(subset=['TEAM_ID','SEASON_ID','STANDINGSDATE'],inplace=True)


ranking_data.drop(['RETURNTOPLAY','LEAGUE_ID'],axis=1).to_csv('./ranking_modified.csv',index=False)
```

# Data Cleaning for teams csv

```python
teams_data=pd.read_csv('./teams.csv')
teams_data
percent_missing_teams= teams_data.isnull().sum() * 100 / len(teams_data)
percent_missing_teams
teams_data.drop(['LEAGUE_ID'],axis=1).to_csv('./teams_modified.csv',index=False)
```
